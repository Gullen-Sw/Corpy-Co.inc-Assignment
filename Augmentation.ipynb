```python
import random
import numpy as np
import os
import cv2
import glob
import math
from PIL import Image
import PIL.ImageOps    

#       .
num_augmented_images_good = 500
num_augmented_images_not_good = 700

file_path_good = '/content/drive/MyDrive/archive/train/normal/good'
file_names_good = os.listdir(file_path_good)
total_origin_image_num_good = len(file_names_good)

file_path_not_good = '/content/drive/MyDrive/archive/train/normal/not-good'
file_names_not_good = os.listdir(file_path_not_good)
total_origin_image_num_not_good = len(file_names_not_good)

#good
for i in range(1, num_augmented_images_good):
    change_picture_index = random.randrange(1, total_origin_image_num_good-1)
    print(change_picture_index)
    print(file_names_good[change_picture_index])
    file_name = file_names_good[change_picture_index]
    
    origin_image_path_good = os.path.join(file_path_good, file_name)
    print(origin_image_path_good)
    image = cv2.imread(origin_image_path_good, cv2.IMREAD_COLOR)

    rows,cols = image.shape[0:2]
    rate1 = random.uniform(0, 1)
    rate2 = random.uniform(1, 2)
    change = cv2.getRotationMatrix2D((cols/2,rows/2),math.pi*rate1, rate2) 
    changed_image = cv2.warpAffine(image, change, (cols, rows))
    cv2.imwrite('/content/drive/MyDrive/archive/train/normal/Aug_good/' + str(i) + '.png', changed_image)

#not-good
for j in range(1, num_augmented_images_not_good):
    change_picture_index = random.randrange(1, total_origin_image_num_not_good-1)
    print(change_picture_index)
    print(file_names_not_good[change_picture_index])
    file_name = file_names_not_good[change_picture_index]
    
    origin_image_path_not_good = os.path.join(file_path_not_good, file_name)
    print(origin_image_path_not_good)
    image = cv2.imread(origin_image_path_not_good, cv2.IMREAD_COLOR)

    rows,cols = image.shape[0:2]
    rate1 = random.uniform(0, 1)
    rate2 = random.uniform(1, 2)
    change = cv2.getRotationMatrix2D((cols/2,rows/2),math.pi*rate1, rate2) 
    changed_image = cv2.warpAffine(image, change, (cols, rows))
    cv2.imwrite('/content/drive/MyDrive/archive/train/normal/Aug_not-good/' + str(j) + '.png', changed_image)
    
```
